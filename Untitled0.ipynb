```python
# Grafo del sector entretenimiento
# Se crea un diccionario que representa un grafo donde cada lugar está conectado con otros
# y cada conexión tiene un peso que puede representar distancia, tiempo o costo.
graph = {
    "Casa": [("Cine",10), ("Parque",5)],  # Desde casa se puede ir al cine o al parque
    "Cine": [("Restaurante",8), ("Centro_Comercial",6)],  # Desde el cine se puede ir a estos dos lugares
    "Parque": [("Bar",7), ("Teatro",12)],  # Desde el parque se puede ir al bar o al teatro
    "Restaurante": [("Discoteca",5)],  # Desde el restaurante se puede ir a la discoteca
    "Centro_Comercial": [("Concierto",15)],  # Desde el centro comercial se llega al concierto
    "Bar": [("Discoteca",6)],  # Desde el bar también se puede ir a la discoteca
    "Teatro": [("Concierto",4)],  # Desde el teatro se puede ir al concierto
    "Discoteca": [("Concierto",3)],  # Desde la discoteca se puede ir al concierto
    "Concierto": []  # El concierto es el destino final, por eso no tiene más conexiones
}

# Se imprimen los nodos del grafo (todos los lugares del sector)
print("NODOS DEL GRAFO")
print(list(graph.keys()))

# Se muestran las conexiones que tiene cada nodo con otros lugares
print("\nCONEXIONES")
for nodo in graph:
    print(nodo,"->",graph[nodo])  # Imprime cada lugar y los lugares a los que se puede ir desde allí
```

DFS, BFS y UCS

```python
# Función que implementa el algoritmo BFS (Breadth First Search)
# Sirve para recorrer el grafo buscando un camino desde un nodo inicial hasta un nodo objetivo
def bfs(graph,start,goal):

    # Conjunto para guardar los nodos que ya fueron visitados
    visited=set()

    # Cola donde se guardan los nodos por explorar
    # También se guarda el camino recorrido y el costo acumulado
    queue=deque([(start,[start],0)])

    # Variable para contar cuántos nodos se han explorado
    explored=0

    # Se guarda el tiempo de inicio para medir cuánto tarda el algoritmo
    inicio=time.time()

    # Mientras existan nodos en la cola se sigue explorando
    while queue:

        # Se extrae el primer elemento de la cola
        node,path,cost=queue.popleft()

        # Se aumenta el contador de nodos explorados
        explored+=1

        # Si el nodo actual es el objetivo, se termina la búsqueda
        if node==goal:
            # Se calcula el tiempo total de ejecución
            tiempo=time.time()-inicio

            # Se devuelve el camino encontrado, el costo, nodos explorados y tiempo
            return path,cost,explored,tiempo

        # Se recorren los vecinos del nodo actual
        for neighbor,weight in graph[node]:

            # Si el vecino aún no ha sido visitado
            if neighbor not in visited:
                # Se marca como visitado
                visited.add(neighbor)

                # Se agrega a la cola con el nuevo camino y el costo actualizado
                queue.append((neighbor,path+[neighbor],cost+weight))
```

```python
# Función que implementa el algoritmo DFS (Depth First Search)
# Este algoritmo recorre el grafo buscando un camino desde un nodo inicial hasta el nodo objetivo
# explorando primero lo más profundo posible en cada rama
def dfs(graph,start,goal):

    # Se crea una pila donde se guardan los nodos por explorar
    # También se guarda el camino recorrido y el costo acumulado
    stack=[(start,[start],0)]

    # Conjunto para guardar los nodos que ya fueron visitados
    visited=set()

    # Variable para contar cuántos nodos se han explorado
    explored=0

    # Se guarda el tiempo de inicio para medir el tiempo de ejecución
    inicio=time.time()

    # Mientras la pila tenga elementos se sigue explorando
    while stack:

        # Se toma el último elemento agregado a la pila
        node,path,cost=stack.pop()

        # Se aumenta el contador de nodos explorados
        explored+=1

        # Si el nodo actual es el objetivo, se termina la búsqueda
        if node==goal:
            # Se calcula el tiempo total de ejecución
            tiempo=time.time()-inicio

            # Se devuelve el camino encontrado, el costo total,
            # los nodos explorados y el tiempo que tardó el algoritmo
            return path,cost,explored,tiempo

        # Si el nodo aún no ha sido visitado
        if node not in visited:
            # Se marca como visitado
            visited.add(node)

            # Se recorren los vecinos del nodo actual
            for neighbor,weight in graph[node]:
                # Se agregan a la pila con el nuevo camino y el costo actualizado
                stack.append((neighbor,path+[neighbor],cost+weight))
```

```python
# Función que implementa el algoritmo UCS (Uniform Cost Search)
# Este algoritmo busca el camino de menor costo desde el nodo inicial hasta el nodo objetivo
def ucs(graph,start,goal):

    # Se crea una cola de prioridad donde se guardan:
    # costo acumulado, nodo actual y el camino recorrido
    pq=[(0,start,[start])]

    # Conjunto para guardar los nodos que ya fueron visitados
    visited=set()

    # Variable para contar cuántos nodos han sido explorados
    explored=0

    # Se guarda el tiempo de inicio para medir el tiempo de ejecución
    inicio=time.time()

    # Mientras haya elementos en la cola de prioridad se sigue explorando
    while pq:

        # Se extrae el elemento con menor costo acumulado
        cost,node,path=heapq.heappop(pq)

        # Se aumenta el contador de nodos explorados
        explored+=1

        # Si el nodo actual es el objetivo se termina la búsqueda
        if node==goal:
            # Se calcula el tiempo total de ejecución
            tiempo=time.time()-inicio

            # Se retorna el camino encontrado, el costo total,
            # los nodos explorados y el tiempo de ejecución
            return path,cost,explored,tiempo

        # Si el nodo aún no ha sido visitado
        if node not in visited:

            # Se marca como visitado
            visited.add(node)

            # Se recorren los vecinos del nodo actual
            for neighbor,weight in graph[node]:
                # Se agregan a la cola de prioridad con el nuevo costo acumulado
                heapq.heappush(pq,(cost+weight,neighbor,path+[neighbor]))
```

```python
# Lista donde se guardarán los resultados de cada algoritmo
resultados=[]

# Se recorren los algoritmos que queremos comparar: BFS, DFS y UCS
for nombre,alg in [("BFS",bfs),("DFS",dfs),("UCS",ucs)]:

    # Se ejecuta cada algoritmo usando el mismo punto de inicio y destino
    path,cost,nodos,tiempo = alg(graph,"Casa","Concierto")

    # Se imprime el nombre del algoritmo que se está ejecutando
    print("\nALGORITMO:",nombre)

    # Se muestra el camino encontrado por el algoritmo
    print("Camino:",path)

    # Se muestra el costo total del camino encontrado
    print("Costo:",cost)

    # Se muestra cuántos nodos fueron explorados durante la búsqueda
    print("Nodos explorados:",nodos)

    # Se muestra el tiempo que tardó el algoritmo en ejecutarse
    print("Tiempo:",tiempo)

    # Se guardan los resultados en la lista para compararlos después
    resultados.append([nombre,nodos,cost,tiempo])

# Se crea una tabla utilizando pandas con los resultados obtenidos
tabla = pd.DataFrame(resultados,
                     columns=["Algoritmo","Nodos Explorados","Costo","Tiempo"])

# Se imprime la tabla comparativa final
print("\nTABLA COMPARATIVA")
tabla
```



Análisis: ¿Cuál algoritmo fue más eficiente y por qué, dado el tipo de grafo de su
sector?

En los resultados obtenidos se observa que el algoritmo más eficiente fue UCS (Uniform Cost Search), ya que tiene en cuenta el costo de cada camino para encontrar la ruta con menor peso total. En el caso del grafo del sector entretenimiento, donde cada conexión tiene un costo asociado (por ejemplo tiempo o distancia entre lugares), UCS logra encontrar una ruta más óptima que BFS o DFS.

Por otro lado, BFS explora los nodos por niveles sin considerar los costos, por lo que puede encontrar un camino con más costo total. DFS tampoco garantiza encontrar la mejor ruta, ya que sigue un camino profundo primero y puede recorrer rutas menos eficientes antes de encontrar la solución.

Por esta razón, para este tipo de grafo donde los caminos tienen diferentes costos, UCS resulta más adecuado porque prioriza siempre el camino de menor costo acumulado.

```python
# Diccionario que contiene los valores heurísticos para cada nodo
# Estos valores representan una estimación del costo que falta para llegar al destino (Concierto)
heuristic={
"Casa":15,
"Cine":10,
"Parque":9,
"Restaurante":6,
"Centro_Comercial":7,
"Bar":6,
"Teatro":4,
"Discoteca":2,
"Concierto":0   # El nodo objetivo tiene heurística 0 porque ya es el destino
}

# Función que implementa el algoritmo A* (A estrella)
# Este algoritmo busca el camino más óptimo combinando el costo real y la heurística
def astar(graph,start,goal,heuristic):

    # Cola de prioridad donde se guarda:
    # f = costo total estimado, nodo actual, camino recorrido y costo real acumulado
    pq=[(0,start,[start],0)]

    # Conjunto para guardar los nodos que ya fueron visitados
    visited=set()

    # Mientras existan nodos en la cola de prioridad se sigue explorando
    while pq:

        # Se extrae el nodo con menor valor de f (prioridad)
        f,node,path,cost=heapq.heappop(pq)

        # Si el nodo actual es el objetivo se devuelve el camino y el costo total
        if node==goal:
            return path,cost

        # Si el nodo aún no ha sido visitado
        if node not in visited:

            # Se marca como visitado
            visited.add(node)

            # Se recorren los vecinos del nodo actual
            for neighbor,weight in graph[node]:

                # g representa el costo real acumulado hasta el vecino
                g=cost+weight

                # h es el valor heurístico estimado desde el vecino hasta el objetivo
                h=heuristic[neighbor]

                # Se agrega el vecino a la cola de prioridad con el valor f = g + h
                heapq.heappush(pq,(g+h,neighbor,path+[neighbor],g))
```

```python
# Se imprime un mensaje para indicar que se va a ejecutar el algoritmo
# antes de hacer cualquier cambio en el grafo o en los valores
print("EJECUCIÓN ANTES DEL CAMBIO")

# Se ejecuta el algoritmo A* desde el nodo "Casa" hasta "Concierto"
# utilizando el grafo y la heurística definida anteriormente
antes = astar(graph,"Casa","Concierto",heuristic)

# Se imprime el camino encontrado por el algoritmo
print("Camino:",antes[0])

# Se imprime el costo total del camino encontrado
print("Costo:",antes[1])
```

• Pregunta extra teórica: ¿Por qué UCS siempre encuentra la solución óptima sin importar
los cambios (si los costos son positivos)?

RTA: El algoritmo UCS (Uniform Cost Search) siempre encuentra la solución óptima cuando los costos son positivos porque explora primero los caminos con menor costo acumulado.

Esto significa que el algoritmo siempre expande el nodo que tiene el costo total más bajo hasta ese momento. Como los costos no pueden ser negativos, una vez que se encuentra el nodo objetivo, se garantiza que no existe otro camino con menor costo que aún no haya sido explorado.

Por esta razón, UCS garantiza encontrar siempre la ruta de menor costo total, incluso si los pesos del grafo cambian, siempre y cuando todos los costos sean positivos.

```python
# Función llamada "buscador" que permite ejecutar diferentes algoritmos de búsqueda
# dependiendo del que el usuario elija
def buscador(origen,destino,algoritmo):

    # Si el usuario selecciona BFS, se ejecuta el algoritmo Breadth First Search
    # usando el origen y destino indicados
    if algoritmo=="BFS":
        return bfs(graph,origen,destino)

    # Si el usuario selecciona DFS, se ejecuta el algoritmo Depth First Search
    if algoritmo=="DFS":
        return dfs(graph,origen,destino)

    # Si el usuario selecciona UCS, se ejecuta el algoritmo Uniform Cost Search
    if algoritmo=="UCS":
        return ucs(graph,origen,destino)

    # Si el usuario selecciona A*, se ejecuta el algoritmo A estrella
    # que utiliza también la heurística definida anteriormente
    if algoritmo=="A*":
        return astar(graph,origen,destino,heuristic)
```

```python
# Función que implementa el algoritmo BFS usando el grafo con múltiples métricas
# Permite elegir qué métrica optimizar (por ejemplo: tiempo, costo o distancia)

def bfs_real(graph,start,goal,metric):

    # Conjunto para guardar los nodos que ya fueron visitados
    visited=set()

    # Cola que guarda: nodo actual, camino recorrido y costo acumulado
    queue=deque([(start,[start],0)])

    # Mientras existan nodos en la cola se sigue explorando
    while queue:

        # Se extrae el primer elemento de la cola (FIFO)
        node,path,cost=queue.popleft()

        # Si el nodo actual es el destino se devuelve el camino y el costo total
        if node==goal:
            return path,cost

        # Se recorren los vecinos del nodo actual
        for neighbor in graph[node]:

            # Se obtiene el peso dependiendo de la métrica seleccionada
            # Puede ser tiempo, costo o distancia
            peso = graph[node][neighbor][metric]

            # Si el vecino aún no ha sido visitado
            if neighbor not in visited:
                # Se marca como visitado
                visited.add(neighbor)

                # Se agrega a la cola con el nuevo camino y el costo acumulado
                queue.append((neighbor,path+[neighbor],cost+peso))
```

```python
# Función que implementa el algoritmo A* utilizando el grafo con múltiples métricas
# Permite elegir qué métrica optimizar (tiempo, costo o distancia)

def astar_real(graph,start,goal,heuristic,metric):

    # Cola de prioridad donde se guarda:
    # f (costo estimado total), nodo actual, camino recorrido y costo real acumulado
    pq=[(0,start,[start],0)]

    # Conjunto para guardar los nodos que ya fueron visitados
    visited=set()

    # Mientras existan nodos en la cola de prioridad se sigue explorando
    while pq:

        # Se extrae el nodo con menor valor de f (prioridad)
        f,node,path,cost = heapq.heappop(pq)

        # Si el nodo actual es el objetivo se devuelve el camino y el costo total
        if node==goal:
            return path,cost

        # Si el nodo aún no ha sido visitado
        if node not in visited:

            # Se marca como visitado
            visited.add(node)

            # Se recorren los vecinos del nodo actual
            for neighbor in graph[node]:

                # Se obtiene el peso dependiendo de la métrica seleccionada
                # Puede ser tiempo, costo o distancia
                peso = graph[node][neighbor][metric]

                # g representa el costo real acumulado hasta el vecino
                g = cost + peso

                # h es el valor heurístico estimado hasta el objetivo
                # Si no existe en el diccionario, se usa 0 por defecto
                h = heuristic.get(neighbor,0)

                # Se agrega el vecino a la cola de prioridad
                # usando la fórmula de A*: f = g + h
                heapq.heappush(pq,(g+h,neighbor,path+[neighbor],g))
```

```python
# Se importa la librería time para poder hacer pausas en la visualización
import time

# Función que permite visualizar cómo se van explorando los nodos del grafo
# empezando desde un nodo inicial
def visualizar_exploracion(graph,start):

    # Conjunto para guardar los nodos que ya fueron visitados
    visitados=set()

    # Cola que se utiliza para recorrer los nodos (similar a BFS)
    cola=[start]

    # Mientras haya nodos en la cola se sigue explorando
    while cola:

        # Se toma el primer nodo de la cola
        nodo = cola.pop(0)

        # Si el nodo aún no ha sido visitado
        if nodo not in visitados:

            # Se marca como visitado
            visitados.add(nodo)

            # Lista que guardará los colores de cada nodo en el grafo
            colores=[]

            # Se recorren todos los nodos del grafo para asignarles un color
            for n in G.nodes():

                # El nodo que se está explorando actualmente se pinta de rojo
                if n == nodo:
                    colores.append("red")

                # Los nodos que ya fueron visitados se pintan de verde
                elif n in visitados:
                    colores.append("green")

                # Los nodos que aún no han sido visitados se pintan de azul claro
                else:
                    colores.append("lightblue")

            # Se dibuja el grafo mostrando los colores de cada nodo
            nx.draw(G,pos,node_color=colores,with_labels=True,node_size=2000)

            # Se muestra la gráfica en pantalla
            plt.show()

            # Se hace una pausa de 1 segundo para que se pueda ver el proceso paso a paso
            time.sleep(1)

            # Se agregan los vecinos del nodo actual a la cola para seguir explorando
            for vecino in graph[nodo]:
                cola.append(vecino)

# Se llama a la función para iniciar la visualización desde el nodo "Casa"
visualizar_exploracion(graph_real,"Casa")
```
